## The `get_comic()` function
Parameters
  - verbose : `bool`
    - by default verbose is set to false. The output is same as the output as the cell above as it uses the `get_latest_comic()` function
  - comic : `str`
    - use string `"latest"` for the latest comic
    - use `"< some integer >"` for getting a specific comic
